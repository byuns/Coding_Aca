```python
import pygame #파이 게임 모듈 임포트
import sys
pygame.init() #파이 게임 초기화
screen = pygame.display.set_mode((300, 300)) #화면 크기 설정
clock = pygame.time.Clock() 
# pygame.display.set_caption("Simple PyGame Example")

#변수
x = 0
y = 0
a = 1
while a: #게임 루프
#     screen.fill((0, 0, 0)) #단색으로 채워 화면 지우기

    #변수 업데이트

#     event = pygame.event.poll() #이벤트 처리
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            a = 0
            break
        if event.type == pygame.MOUSEBUTTONDOWN and event.button == 1.:
            print("마우스 왼쪽 클릭")
            print(event.button)
            print(pygame.mouse.get_pos())
      
        if event.type == pygame.MOUSEBUTTONUP:
            print("마우스 누르고 떼다.")
            print(event.button)

        if event.type == pygame.KEYDOWN:
            print("키 누름")
            if event.key == pygame.K_SPACE:
                print("스페이스")
            elif event.key == pygame.K_LEFT:
                x -= 1
            elif event.key == pygame.K_RIGHT:
                x += 1
            elif event.key == pygame.K_UP:
                y += 1
            elif event.key == pygame.K_DOWN:
                y -= 1
            print('(',x,',',y,')')
        

    #화면 그리기

#     pygame.display.update() #모든 화면 그리기 업데이트
#     clock.tick(30) #30 FPS (초당 프레임 수) 를 위한 딜레이 추가, 딜레이 시간이 아닌 목표로 하는 FPS 값

pygame.quit()    
```

```python
import os
import time
import pygame
import random
from pygame.locals import *
import time

pygame.init()

screen = pygame.display.set_mode((1000,700))

pygame.display.set_caption("Moving")

clock = pygame.time.Clock()
background = pygame.image.load(os.path.join("배경.png"))
img_scale = pygame.transform.scale(background,(1000,700))

ship = pygame.image.load(os.path.join("어몽어스.png"))
ship = pygame.transform.scale(ship,(100,100))

x = 0
y = 0
a = 1
xs = 0.1
ys = 0.1

DOWN = False
UP = False
LEFT = False
RIGHT = False


while a:
    for event in pygame.event.get():
        if event.type == KEYDOWN:
            if event.key == K_LEFT or event.key == ord('a'):
                LEFT = True
                RIGHT = False
                
            if event.key == K_RIGHT or event.key == ord('d'):
                RIGHT = True
                LEFT = False
                
            if event.key == K_UP or event.key == ord('w'):
                UP = True
                DOWN = False
               
            if event.key == K_DOWN or event.key == ord('s'):
                DOWN = True
                UP = False
        if event.type == KEYUP:
            if event.key == K_LEFT or event.key == ord('a'):
                LEFT = False
            if event.key == K_RIGHT or event.key == ord('d'):
                RIGHT = False
            if event.key == K_UP or event.key == ord('w'):
                UP = False
            if event.key == K_DOWN or event.key == ord('s'):
                DOWN = False
        elif event.type == pygame.QUIT:
            a = 0
            break
    if LEFT:
        x -= xs
    if RIGHT:
        x += xs
    if UP:
        y -= ys
    if DOWN:
        y += ys
    screen.blit(img_scale,(0,0))
    screen.blit(ship,(x,y)) 
    pygame.display.update()
pygame.quit()
    
```
