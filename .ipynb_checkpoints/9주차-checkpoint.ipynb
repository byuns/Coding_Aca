```python
import random

tmp = []
for i in range(5):
    tmp.append(random.randint(1,9))
    
print(tmp)
avg = sum(tmp)/len(tmp)
down = []
up = []
for i in range(len(tmp)):
    if tmp[i]< avg:
        down.append(tmp[i])
    else:
        up.append(tmp[i])
print(avg,down,up)
```

```python
# 2,3,5,7,11

a = []

for i in range(2,100):
    a.append(i)

pri_num = []

for i in range(len(a)):
    for j in range(2,a[i]):
        if a[i]%j != 0:
            pri_num.append(a[i])
pri_num
```
