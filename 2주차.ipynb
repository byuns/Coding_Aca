```python
import random
while True:
    a = input("오늘의 번호는?")
    num = a.split()
    if len(num) != 2:
        continue
    else:
        if num[0] == num[1]:
            print("숫자 입력을 제대로 해주세요")
            continue
        else:
            if int(num[0]) > 0 and int(num[0]) <= 10 and int(num[1]) > 0 and int(num[1]) <= 10:
                break
            else:
                print("숫자 입력을 제대로 해주세요")
                continue

count_1 = 0
count_2 = 0
count_3 = 0
count_zero = 0
                
for i in range(1,101):
    while True:
        num_1 = random.randint(1,10)
        num_2 = random.randint(1,10)
        if num_1 != num_2:
            break
    print(i,"번 째 구매한 로또 번호는 ",num_1,num_2,"이다.")
    if num_1 == int(num[0]) and num_2 == int(num[1]):
        print("축하합니다. 1등에 당첨되었습니다.")
        count_1 += 1
    else:
        if num_1 == int(num[1]) and num_2 == int(num[0]):
            print("축하합니다. 2등에 당첨되었습니다.")
            count_2 += 1
        elif num_1 == int(num[0]) and num_2 != int(num[1]):
            print("축하합니다. 3등에 당첨되었습니다.")
            count_3 += 1
        elif num_1 == int(num[1]) and num_2 != int(num[0]):
            print("축하합니다. 3등에 당첨되었습니다.")
            count_3 += 1
        elif num_1 != int(num[1]) and num_2 == int(num[0]):
            print("축하합니다. 3등에 당첨되었습니다.")
            count_3 += 1
        elif num_1 != int(num[0]) and num_2 == int(num[1]):
            print("축하합니다. 3등에 당첨되었습니다.")
            count_3 +=1
        else:
            print("꽝,아쉽지만 다음 기회에..")
            count_zero +=1
            
print("1등 : ",count_1,"회")
print("2등 : ",count_2,"회")
print("3등 : ",count_3,"회")
print("꽝 : ",count_zero,"회")
        
```
