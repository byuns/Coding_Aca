```python
import os
import time
import pygame
import random
from pygame.locals import *
import time

pygame.init()

screen = pygame.display.set_mode((1000,700))

pygame.display.set_caption("Moving")

clock = pygame.time.Clock()
background = pygame.image.load(os.path.join("흰배경.jpg"))
img_scale = pygame.transform.scale(background,(1000,700))

line_1 = pygame.image.load(os.path.join("선.png"))
line_1 = pygame.transform.scale(line_1,(5,500))

line_2 = pygame.image.load(os.path.join("선.png"))
line_2 = pygame.transform.scale(line_2,(5,500))

ship = pygame.image.load(os.path.join("어몽어스.png"))
ship = pygame.transform.scale(ship,(100,100))

# en = pygame.image.load("search.pstatic-removebg-preview (9).png")
# en = pygame.transform.scale(en,(100,100))

x = 0
y = 0
a = 1

enx = 0
eny = 0
ens = 0.5
count = 0

lx_1 = 300
ly_1 = 200
lx_2 = 300
ly_2 = 200

speed = 0.5
DOWN = False
UP = False
LEFT = False
RIGHT = False


while a:
    for event in pygame.event.get():
        if event.type == KEYDOWN:
            if event.key == K_LEFT or event.key == ord('a'):
                LEFT = True
                RIGHT = False
                
            if event.key == K_RIGHT or event.key == ord('d'):
                RIGHT = True
                LEFT = False
                
            if event.key == K_UP or event.key == ord('w'):
                UP = True
                DOWN = False
               
            if event.key == K_DOWN or event.key == ord('s'):
                DOWN = True
                UP = False
        if event.type == KEYUP:
            if event.key == K_LEFT or event.key == ord('a'):
                LEFT = False
            if event.key == K_RIGHT or event.key == ord('d'):
                RIGHT = False
            if event.key == K_UP or event.key == ord('w'):
                UP = False
            if event.key == K_DOWN or event.key == ord('s'):
                DOWN = False
        elif event.type == pygame.QUIT:
            a = 0
            break
    if LEFT:
        x -= speed
    if RIGHT:
        x += speed
    if UP:
        y -= speed
    if DOWN:
        y += speed
    character_rect = ship.get_rect()
    character_rect.left = x
    character_rect.top = y
    
#     enemy_rect = en.get_rect()
#     enemy_rect.left = enx
#     enemy_rect.top = eny

    ee_1 = line_1.get_rect()
    ee_1.left = lx_1
    ee_1.top = ly_1
    
    ee_2 = line_2.get_rect()
    ee_2.left = lx_2
    ee_2.top = ly_2
    
    if character_rect.colliderect(ee_1):
        x = 0
        y = 0
    if character_rect.colliderect(ee_2):
        x = 0
        y = 0
    
#     if character_rect.colliderect(enemy_rect):
#         ens += 0.1
#         enx = 0 
#         eny = random.randint(1,650)
#         count += 1
    
#     enx += ens
#     if enx > 1000:
#         enx = 0
#         eny = random.randint(1,650)
        
#     if count == 10:
#         break
    
    
    screen.blit(img_scale,(0,0))
    screen.blit(line_1,(lx_1,ly_1))
    screen.blit(line_2,(lx_2,ly_2))
    screen.blit(ship,(x,y)) 
#     screen.blit(en,(enx,eny))
    pygame.display.update()
pygame.quit()
    
```
